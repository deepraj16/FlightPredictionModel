### 4.Data Preprocessing

```python
dt_cols = ["date_of_journey", "dep_time", "arrival_time"]

num_cols = ["duration", "total_stops"]

cat_cols = [col for col in X_train.columns if (col not in dt_cols) and (col not in num_cols)]
```

```python
cat_cols
```

```python
num_transformer = Pipeline(steps=[
	("imputer", SimpleImputer(strategy="median")),
	("scaler", StandardScaler())
])

cat_transformer = Pipeline(steps=[
	("imputer", SimpleImputer(strategy="most_frequent")),
	("encoder", OneHotEncoder(sparse_output=False, handle_unknown="ignore"))
])

doj_transformer = Pipeline(steps=[
	("imputer", SimpleImputer(strategy="most_frequent")),
	("extractor", DatetimeFeatures(features_to_extract=["month", "week", "day_of_week", "day_of_month"], format="mixed")),
	("scaler", StandardScaler())
])

time_transformer = Pipeline(steps=[
	("imputer", SimpleImputer(strategy="most_frequent")),
	("extractor", DatetimeFeatures(features_to_extract=["hour", "minute"], format="mixed")),
	("scaler", StandardScaler())
])
```

```python
preprocessor = ColumnTransformer(transformers=[
	("num", num_transformer, num_cols),
	("cat", cat_transformer, cat_cols),
	("doj", doj_transformer, ["date_of_journey"]),
	("time", time_transformer, ["dep_time", "arrival_time"])
])
```

```python
preprocessor.fit_transform(X_train)
```

```python
preprocessor.fit_transform(X_train).shape
```

# 5.model Selection


```python
algorithms ={
    "Linear Regression ": LinearRegression() , 
    "Support Vector Machine": SVR(), 
    "Random Forest":RandomForestRegressor(n_estimators=10) , 
    "XG Boost"  :XGBRegressor(n_estimators=10)
}
```

```python
data =pd.concat([train_df,val_df],axis=0)
X_data,y_data=split_data(data)
print(X_data.shape) 
print(y_data.shape)
```

```python
def plot_learning_curver(name,algorithim,figsize=(12,4)): 
    model =Pipeline( steps=[
        ("pre" ,preprocessor ), 
        ("alg" , algorithim)
    ])
    train_size,train_score,test_scores = learning_curve(
        estimator=model, 
        X= X_data, 
        y=y_data, 
        cv=3, 
        scoring="r2", 
        n_jobs=1 , 
        random_state=42
    )
    mean_train_score =np.mean(train_score,axis=1)
    std_train_score =np.std(train_score,axis=1)
    train_label =f"{mean_train_score[-1]:.2f}+/- {std_train_score[-1]:.2f}"
    
    
    mean_test_score =np.mean(test_scores,axis=1)
    std_test_score =np.std(test_scores,axis=1)
    fig,ax =plt.subplots(figsize=figsize)
    
    test_label =f"{mean_test_score[-1]:.2f}+/- {std_test_score[-1]:.2f}"
    
    plot_curves(
        train_size,mean_train_score,std_train_score,f"Trian{train_label}",ax
    )
    
    
    plot_curves(
        train_size,mean_test_score,std_test_score,f"Trian{test_label}",ax
    )
    
    ax.set(xlabel="traingset size",ylabel="R-square",title=name) 
    ax.legend(loc="lower right")
    plt.show()
# traing folds Test folds 
```

```python
def plot_curves(sizes,mean_scores,std_score,label,ax):
    ax.plot(
        sizes, 
        mean_scores, 
        marker="o", 
        label=label
    )
    ax.fill_between(
        x=sizes, 
        y1=mean_scores -std_score, 
        y2=mean_scores+std_score, 
        alpha=0.5
    )
```

```python
for name,alg in algorithms.items(): 
    plot_learning_curver(name,alg)
```

# 6.model Training

```python
model = Pipeline( steps=[
    ("pre",preprocessor), 
    ('rf',RandomForestRegressor(n_estimators=10))
])
```

```python
model.fit(X_data,y_data)
```

## 7model Evalution

```python
def evaluate_model(X,y): 
    y_pred =model.predict(X) 
    return r2_score(y,y_pred)
```
